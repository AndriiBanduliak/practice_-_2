```python
# Импорт библиотек
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Наборы данных
X1 = np.array([30,30,40, 40])
Y1 = np.array([37, 47, 50, 60])

X2 = np.array([30,30,40, 40, 20, 20, 50, 50])
Y2 = np.array([37, 47, 50, 60, 25, 35, 62, 72])

X3 = np.array([30,30,40, 40, 20, 20, 50, 50, 10, 10, 60, 60])
Y3 = np.array([37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84])

# Функция для построения графика рассеивания и модели линейной регрессии
def plot_regression_line(x, y, title):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    line = slope * x + intercept
    plt.plot(x, y, 'o', x, line)
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    print("Коэффициенты линейной регрессии: slope = {:.2f}, intercept = {:.2f}".format(slope, intercept))
    print("Коэффициент детерминации: r-squared = {:.2f}".format(r_value ** 2))

# Построение графика рассеивания и модели линейной регрессии для первого набора данных
plot_regression_line(X1, Y1, "Набор данных 1")

# Построение графика рассеивания и модели линейной регрессии для второго набора данных
plot_regression_line(X2, Y2, "Набор данных 2")

# Построение графика рассеивания и модели линейной регрессии для третьего набора данных
plot_regression_line(X3, Y3, "Набор данных 3")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# задаем данные
df1 = pd.DataFrame({'X': [30, 30, 40, 40], 'Y': [37, 47, 50, 60]})
df2 = pd.DataFrame({'X': [30, 30, 40, 40, 20, 20, 50, 50], 'Y': [37, 47, 50, 60, 25, 35, 62, 72]})
df3 = pd.DataFrame({'X': [30, 30, 40, 40, 20, 20, 50, 50, 10, 10, 60, 60], 'Y': [37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84]})

# построение графиков и расчет коэффициентов
for df in [df1, df2, df3]:
    x = df['X']
    y = df['Y']
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    r_squared = r_value ** 2
    plt.scatter(x, y)
    plt.plot(x, intercept + slope * x, 'r')
    plt.show()
    print(f'Коэффициенты для линии регрессии: наклон={slope}, пересечение={intercept}')
    print(f'Коэффициент детерминации: {r_squared:.2f}\n')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def plot_regression(x, y):
    # создание модели линейной регрессии
    model = LinearRegression()
    # преобразование массивов данных в нужный формат
    x = x.reshape((-1, 1))
    # обучение модели на данных
    model.fit(x, y)
    # предсказание значений y на основе обученной модели
    y_pred = model.predict(x)
    # вывод коэффициентов регрессии
    print('Коэффициенты регрессии:', model.coef_)
    print('Пересечение:', model.intercept_)
    # вывод коэффициента детерминации
    print('Коэффициент детерминации:', model.score(x, y))
    # построение графика
    plt.scatter(x, y)
    plt.plot(x, y_pred, color='red')
    plt.show()

```

```python
# данные из первого набора
x1 = np.array([10,8, 13, 9,11,14, 6,4,12, 7,5])
y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
plot_regression(x1, y1)


```

```python
# данные из второго набора
x2 = np.array([10,8, 13, 9,11,14, 6,4,12, 7,5])
y2 = np.array([9.14, 8.14, 8.74,8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74])
plot_regression(x2, y2)
```
